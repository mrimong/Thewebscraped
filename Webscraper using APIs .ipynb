### Using APIs

```python
#IMPORTING LIBRARIES
import requests
from bs4 import BeautifulSoup
import json
import time
#FOR SQL TABLE CREATION
import mysql.connector
import warnings
import codecs
```

## The code will go to https://api.github.com/repos/apache/hadoop/contributors . This is the Apache Hadoop Github Repo's contributorsâ€™ endpoint. Extract the JSON corresponding to the first 100 contributors from this API. (The API request is a GET request and the variable name that handles the items per page is "per_page")

```python
headers = {'User-agent': 'Mozilla/5.0'}
#I HAVE LEVERAGED THE 'per_page' TO CONTROL THE NUMBER OF ITEMS BEING DISPLAYED IN A PAGE. I HAVE REFERRED TO 
#THE "https://api.github.com/" page
url='https://api.github.com/repos/apache/hadoop/contributors?per_page=100'
page=requests.get(url,headers)
doc=BeautifulSoup(page.content,'html.parser')
#to parse a valid JSON string and convert it into a Python Dictionary.
json_dict=json.loads(str(doc))
```

## Code  accesses user information and extracts "login", "id", "location", "email", "hireable", "bio", "twitter_username", "public_repos", "public_gists", "followers", "following", "created_at"

```python
json_dict[0]
#BASED ON THE OUTPUT, I CAN SEE THAT EVERY CONTRIBUTOR HAS AN UNIQUE URL THAT IS A LINK TO THEIR PROFILE 
```

```python
#CREATING AN EMPTY LIST TO STORE ALL THE URLS OF THE CONTRIBUTORS AND APPENDING THE URLS TO THE LIST
contributor_url=[]
for i in range(len(json_dict)):
    contributor_url.append(json_dict[i]['url'])
```

```python
#DISPLAYING ALL THE 100 URLS OF THE CONTRIBUTORS
contributor_url
```

```python
#CREATING EMPTY LISTS FOR THE DIFFERENT VARIABLES THAT WE NEED
login = []
id1= []
location = []
email = []
hireable = []
bio = []
twitter_username = []
public_repos = []
public_gists = []
followers = []
following = []
created_at = []
#SINCE I AM GETTING BLOCKED/RATE LIMITED BY THE URL, I HAVE ONLY CONSIDERED 15 CONTRIBUTORS FOR THE CODE.
#AS PER PIAZZA POST: https://piazza.com/class/lqya6hx57zd7c/post/171

#APPENDING THE VALUES TO THE EMPTY LISTS FOR THE CONTIRBUTORS
for i in range(15):    
    url_q3=contributor_url[i]
    time.sleep(10)
    page_q3=requests.get(url_q3,headers)
    doc_q3=BeautifulSoup(page_q3.content,'html.parser')
    #to parse a valid JSON string and convert it into a Python Dictionary.
    json_dict_q3=json.loads(str(doc_q3))

    login.append(json_dict_q3['login'])
    print("Appended for user",json_dict_q3['login'])
    id1.append(json_dict_q3['id'])
    location.append(json_dict_q3['location'])
    email.append(json_dict_q3['email'])
    hireable.append(json_dict_q3['hireable'])
    bio.append(json_dict_q3['bio'])
    twitter_username.append(json_dict_q3['twitter_username'])
    public_repos.append(json_dict_q3['public_repos'])
    public_gists.append(json_dict_q3['public_gists'])
    followers.append(json_dict_q3['followers'])
    following.append(json_dict_q3['following'])
    created_at.append(json_dict_q3['created_at'])

```

```python
#DISPLAYING THE INFORMATION OF THE 100 CONTRIBUTOR PROFILES
for i in range(15):
    print("CONTRIBUTOR NUMBER->",i+1
            ,"\nLogin",login[i]
            ,"\nID:",id1[i]
            ,"\nLocation:", location[i]
            ,"\nEmail:", email[i]
            ,"\nHireable:", hireable[i]
            ,"\nBio:", bio[i]
            ,"\nTwitter Username:", twitter_username[i]
            ,"\nPublic Repos:", public_repos[i]
            ,"\nPublic Gists:", public_gists[i]
            ,"\nFollowers:", followers[i]
            ,"\nFollowing:", following[i]
            ,"\nCreated At:", created_at[i]
            ,"\n----------------------------------------------------------")
```

## Creating a SQL database + table, and stores all the information obtained above.

```python
#REPLACING THE NONE ELEMENTS IN THE LIST FOR SQL INSERTION. I AM REPLACING THE NONE ELEMENTS WITH A BLANK ''
id1=[s if s is not None else '' for s in id1]
login=[s if s is not None else '' for s in login]
location= [s if s is not None else '' for s in location]
email= [s if s is not None else '' for s in email]
hireable= [s if s is not None else '' for s in hireable]
bio= [s if s is not None else '' for s in bio]
twitter_username= [s if s is not None else '' for s in twitter_username]
public_repos= [s if s is not None else '' for s in public_repos]
public_gists= [s if s is not None else '' for s in public_gists]
followers= [s if s is not None else '' for s in followers]
following= [s if s is not None else '' for s in following]
created_at= [s if s is not None else '' for s in created_at]
```

```python
#ignore warnings
warnings.filterwarnings("ignore")
SQL_DB = "ucdavis"
#CREATING FUNCTION TO CREATE DATABASE AND SQL TABLE
def create_sql_table(SQL_TABLE_URBAN, SQL_TABLE_URBAN_DEF):
    try:
        #connect to server
        conn = mysql.connector.connect(host='localhost',
                                       user='root',
                                       password='qwerty123')
        cursor = conn.cursor()
        
        query = "CREATE DATABASE IF NOT EXISTS " + SQL_DB
        print(query)
        cursor.execute(query);
        
        query = "CREATE TABLE IF NOT EXISTS " + SQL_DB + "." + SQL_TABLE_URBAN + " " + SQL_TABLE_URBAN_DEF + ";";
        print(query)
        cursor.execute(query);
        cursor.close()
        conn.close()
        return

    except IOError as e:
        print(e)
```

```python
#OUR TABLE NAME IS github
SQL_TABLE_URBAN = "github"
#DEFINING THE TABLE WITH THE DATA TYPES AND SIZES
SQL_TABLE_URBAN_DEF = "(" + \
        "id INT NOT NULL PRIMARY KEY" + \
        ",login VARCHAR(50)" + \
        ",location VARCHAR(100)" + \
        ",email VARCHAR(100)" + \
        ",hireable VARCHAR(100)" + \
        ",bio VARCHAR(1000)" + \
        ",twitter_username VARCHAR(100)" + \
        ",public_repos INT" + \
        ",public_gists INT" + \
        ",followers INT" + \
        ",following INT" + \
        ",created_at VARCHAR(100)" + \
        ")"

create_sql_table(SQL_TABLE_URBAN, SQL_TABLE_URBAN_DEF)



try:
    # connect to server
    conn = mysql.connector.connect(host='localhost',
                                   database='ucdavis',
                                   user='root',
                                   password='qwerty123')

    cursor = conn.cursor()

 
    # INSERTION CODE
    parameterized_stmt = "INSERT INTO " + SQL_TABLE_URBAN + "(id, login, location, email, hireable, bio, twitter_username, public_repos, public_gists, followers, following, created_at) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);"
    #SINCE I STORED MY VALUES IN LISTS, I AM INSERTING EACH ITEM OF THE LIST TO THE TABLE
    for i in range(15):
        #parameterized version
        cursor.execute(parameterized_stmt, (id1[i],login[i], location[i], email[i], hireable[i], bio[i]
                                            , twitter_username[i], public_repos[i], public_gists[i]
                                            , followers[i], following[i], created_at[i]))
        print(i," Done") #PRINTING A SUCCESSFUL INSERTION MESSAGE 
        
    conn.commit()
    cursor.close()
    conn.close()
    
except IOError as e:
    print(e)
```
