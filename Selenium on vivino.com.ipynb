### Using Selenium to remote control a browser to [access] https://www.vivino.com/US-CA/en/, [clicks] on
“Wines” in the upper left hand corner, and displays all “Red” wines (only “Red”, “White” needs to be
[deactivated]). [Select] “Any rating”

```python
#Setting the url
url_new='https://www.vivino.com/US-CA/en/'
#Creating a new browser window using selenium
driver= webdriver.Chrome() 
time.sleep(5)
driver.get(url_new)
time.sleep(5)
#I used the text of "Wines" menu to clcik on wines
wines=driver.find_element(By.XPATH,'//span[text()="Wines"]')
wines.click()
time.sleep(5)
#I used the text of "White" option to clcik on white and to remove it as a filter
white=driver.find_element(By.XPATH,'//span[text()="White"]')
white.click()
time.sleep(3)
#I used the text of "Rose" option to clcik on rose and to remove it as a filter
rose=driver.find_element(By.XPATH,'//span[text()="Rosé"]')
rose.click()
time.sleep(3)
#I used the text of "Sparkling" option to clcik on sparkling and to remove it as a filter
spark=driver.find_element(By.XPATH,'//span[text()="Sparkling"]')
spark.click()
time.sleep(3)
#I used the text of "any rating" option to clcik on any rating and to select it as a filter
rating=driver.find_element(By.XPATH,'//div[text()="Any rating"]')
rating.click()
time.sleep(3)

time.sleep(5)
driver.quit()
```

###
For the wine IDs 1, 11, and 128, [using] Selenium to access the pages https://www.vivino.com/USCA/en/w/[ID]. Leaving some pause between page loads. [Printing] the ID to terminal. [Reading] the URL of the 
resulting page and [print] it to terminal. [Checking] whether it exists (404 error or not) and [print] to 
terminal “404 error” or “no 404 error”. Lastly, checking if the resulting page has an ID different to the one 
you tried to access and [printing] to terminal “not forwarded” or “forwarded to [ID]”.

```python
#Setting the wine id
wine_id=[1,11,128]
#URL
url="https://www.vivino.com/US-CA/en/w/"
#Loop for each wine id
for i in wine_id:

    print("Wine ID:",i)
    time.sleep(5)
    url_c=url+str(i)
    print("URL:",url_c)
    #Setting the url
    #Creating a new browser window using selenium
    driver= webdriver.Chrome() 
    time.sleep(5)
    driver.get(url_c)
    time.sleep(5)
    try:
       #I noticed that all the pages that do not have a relevant id contain an error page which has the below class
        error=driver.find_element(By.CLASS_NAME, "error-page-text-container")
        print("404 error")
        print("not forwarded")
        #Error pages are not forwarded hence displaying the same
        
    except:
        print("no 404 error")
        url_new=driver.current_url
        #to check the url if it was forwarded or not, I wanted to find the current url of the page
        #Referred to: https://www.geeksforgeeks.org/how-to-get-current_url-using-selenium-in-python/
        url_id=re.findall(r'.*/w/(.*)',url_new)
        #This finds the regex for the wine id after w/ in the URL
        if url_id[0]==str(i):
            print("not forwarded")
        else:
            print("forwarded to",url_id[0])
    driver.quit()
    print("\n")
    time.sleep(3)
```

### Using Selenium to access the wine ID 1695288 (exists, is not forwarded, and is Lokoya’s expensive 
Cabernet Sauvignon). [Scrolling] to the bottom of the page. In Selenium, [scrolling] can be achieved by 
[pressing] the page-down button about 10 times.

```python
#Settign the URL and opening a new wiindow
urld="https://www.vivino.com/US-CA/en/w/1695288"

driver=webdriver.Chrome()
time.sleep(5)
driver.get(urld)
time.sleep(5)
# Referred to : https://www.geeksforgeeks.org/action-chains-in-selenium-python/
#To simulate scrolling I used ActionChains 
actions = ActionChains(driver)
i=1
while i<11:
    actions.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
    time.sleep(5)
    print("Scrolling done for",i,"th time")
    i+=1
driver.quit()
```

```python
#Reading file from local system
with open("vivino_id_1695288.html", "r") as file:
    html_content = file.read()

#Parsing the file and displaying all the content
soup = BeautifulSoup(html_content, "html.parser")
print(soup.prettify())
```

```python
#finding wine id-> "wine_id":1695288,
#WINES
#using regex i found wine id is stored in a json format. I used regex to find the numerical ID for wines
wine_id=re.findall(r'"wine_id":(\d+)',str(soup))[0]
#Winery name is stored in <A> tag under class winery
winery=soup.find_all("a",{"class": "winery"})[0].text.strip()
#the wine names are stored under class vintage
wine_name=soup.find_all("span",{"class": "vintage"})[0].text.strip()
#Grapes are stored in a table with each table data having its own class
grapes=soup.find_all("td",{"class":"wineFacts__fact--3BAsi"})[1].text
#Badges can be found in the righcolumn class for all the wines. 
badgeclass=soup.find_all("div",{"class":"rightColumn"})
#print(badges)
badges_list=[]
for i in badges_list:
    badges_list.append(i.text.strip())
#I noticed some repitions of badges so I am using unique() to keep only unique values
badges_list=np.unique(badges_list)
#Region, Winestlye, Acloclhol etc are stored under WineFacts class which is also a table data element
region=soup.find_all("td",{"class":"wineFacts__fact--3BAsi"})[2].text
winestyle=soup.find_all("td",{"class":"wineFacts__fact--3BAsi"})[3].text
alcohol=soup.find_all("td",{"class":"wineFacts__fact--3BAsi"})[4].text
allergens=soup.find_all("td",{"class":"wineFacts__fact--3BAsi"})[5].text
#Ratings are stored under vivinoRating class
avg_rating=soup.find_all("div",{"class":"vivinoRating_averageValue__uDdPM"})[0].text
num_rating=soup.find_all("div",{"class":"vivinoRating_caption__xL84P"})[0].text
#Price is under CurrentPrice class
price=soup.find_all("span",{"class":"purchaseAvailability__currentPrice--3mO4u"})[0].text
#The different food items are stored under a class with name foodpairiing
food=soup.find_all("div",{"class":"foodPairing__foodContainer--1bvxM"})[0].text
date_of_download = str(date.today())

#REVIEWS
#Finding the user id under review section
userid=soup.find_all("div",{"class":"card__card--2R5Wh communityReview__card--1a-k4"})[0]
#Ratng given by the user
rating_0=soup.find_all("span",{"class":"userRating_ratingNumber__cMtKU"})
ratings=[]
#Rating is usually 2-digit seperated by a . in the middle
for i in rating_0:
    r=re.findall(r'([0-9].[0-9])',str(i))
    ratings.append(r[0])
    
username=[]
num_userreviews=[]
#Using the regex I find the username and num of user reviews 
username_0=soup.find_all("a"
,{"class":"anchor_anchor__m8Qi- userAlias_userAlias__ztmrT undefined communityReview__userAlias--1wUOM anchor_baseLink__O8bvu"})
for n in range(len(username_0)):
    for i in username_0[n]:
        username.append(re.findall(r'(.+?)\([\d]',str(i))[0])
        num_userreviews.append(re.findall(r'\((\d+)\s+ratin',str(i))[0])
#To store the review text i used the communityreview class
review_0=soup.find_all("span",{"class":"communityReview__reviewText--2bfLj"})
review=[]
for i in review_0:
    review.append(i.text)
#Like and COmments are in like/comment buttons under div tag
like_0=soup.find_all("div",{"class":"likeButton__likeCount--1stJS"})
comment_0=soup.find_all("div",{"class":"commentsButton__commentsCount--3CoCn"})

like=[]
comment=[]
#Appendingthem to list
for i,j in zip(like_0,comment_0):
    like.append(i.text)
    comment.append(j.text)
#Vintage year is found above the "add to cart" option    
vintage_year=soup.find_all("span",{"class":"css-14ngluw-componentChildren"})[1].text
#USER ID is found with the href of each review and I used regex to find the review
userid_0=soup.find_all("a"
,{"class":"anchor_anchor__m8Qi- reviewAnchor__anchor--2NKFw communityReview__reviewContent--3xA5s"})
userid=[]
for i in userid_0:
    userid.append(re.findall(r'reviews/(.*)',i['href'])[0])
```

```python
#LINKS  
#Link of the other wines at the bottom are stored in winecards. I used these href to get wine id
other_link=soup.find_all("a",{"class":"anchor_anchor__m8Qi- wineCard__cardLink--3F_uB"})
other_links=[]
for i in other_link:
    other_links.append(re.findall(r'[w]/(.*).price',i['href'])[0])
other_links

#TASTE
#THe taste percentage is stored in a progress bar 
taste=soup.find_all("span",{"class":"indicatorBar__progress--3aXLX"})
light_bold=[]
smooth_dry=[]
dry_sweet=[]
soft_acidic=[]
#All the taste elemtns have the style left % whcih determines the idnicator of the particular taste like sweet 85%
#I used this and a comparison to determine sweetness, dryness, acidity etc
for cnt in range(len(taste)):
    s=re.findall(r'left: (.*)%;',str(taste[cnt]))
    #light bold
    if(float(s[0])>0.0 and cnt==0):
        light_bold.append(1)
    else:
        light_bold.append(0)
    
    #smooth dry
    if(float(s[0])>0.0 and cnt==1):
        smooth_dry.append(1)
    else:
        smooth_dry.append(0)

    #dry sweet
    if(float(s[0])>0.0 and cnt==1):
        dry_sweet.append(1)
    else:
        dry_sweet.append(0)
    
    #Soft acidic
    if(float(s[0])>0.0 and cnt==1):
        soft_acidic.append(1)
    else:
        soft_acidic.append(0)
    
```

### Creating a mongoDB to save and store the wine data

```python
#Setting up mongoDB connection 
client = MongoClient("localhost", 27017)
#Creating a database 
db=client["vivino"]

#Creating a collection of ip-addresses in the database
collection1= db["wines"]
collection2= db["reviews"]
collection3= db["links"]
collection4= db["taste"]

#insert information in Wines collection
document = {
    "wine_id": wine_id,
    "winery": winery,
    "wine_name": wine_name,
    "grapes": grapes,
    "badges_list": badges_list.tolist(),
    "region": region,
    "winestyle": winestyle,
    "alcohol": alcohol,
    "allergens": allergens,
    "avg_rating": avg_rating,
    "num_rating": num_rating,
    "price": price,
    "food": food,
    "date_of_download": date_of_download
}
collection1.insert_one(document)

#Insert information in the reviews collection
for i in range(len(userid)):
    document = {
        "userid": userid[i],
        "username": username[i],
        "num_userreviews": num_userreviews[i],
        "ratings": ratings[i],
        "review": review[i],
        "like": like[i],
        "comment": comment[i],
        "vintage_year":vintage_year
    }
    collection2.insert_one(document)

#Inserting the information in the links collection
for i in range(len(other_links)):
    document = {"id": other_links[i]}
    collection3.insert_one(document)
    
#Inserting the information in the taste collection
for i in range(len(taste)):
    document = {"light_bold": light_bold[i]
                ,"smooth_dry": smooth_dry[i]
                ,"dry_sweet": dry_sweet[i]
                ,"soft_acidic": soft_acidic[i]}
    collection4.insert_one(document)

#Printing the information in the collections
query_result = collection1.find({})
for doc in query_result:
    print(doc)
    print("/n")
    
query_result = collection2.find({})
for doc in query_result:
    print(doc)
    print("/n")
    
query_result = collection3.find({})
for doc in query_result:
    print(doc)
    print("/n")

query_result = collection4.find({})
for doc in query_result:
    print(doc)
    print("/n")
    
#Closing the connection
client.close()
```
